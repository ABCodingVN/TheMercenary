```python
from bs4 import BeautifulSoup
import requests
import json
import csv
import warnings
warnings.filterwarnings("ignore")
import pandas as pd


typeStock="sach"
laptop_page_url = "https://tiki.vn/laptop/c8095?src=c.8095.hamburger_menu_fly_out_banner&_lc=&page={}"
sach_url ="https://tiki.vn/search?q=s%C3%A1ch&page={}"
thoi_trang_url="https://tiki.vn/search?q=th%E1%BB%9Di+trang&page={}"
hang_tieu_dung="https://tiki.vn/search?q=gia+v%E1%BB%8B&page={}"
main_url=sach_url
product_url = "https://tiki.vn/api/v2/products/{}"
product_id_file = f"{typeStock}-id.txt"
product_data_file = f"{typeStock}.txt"
product_file = f"{typeStock}.csv"
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}

def crawl_product_id():
    product_list = []
    i = 1
    while (i<40):
        print("Crawl page: ", i)
        response = requests.get(main_url.format(i),headers=headers)
        parser = BeautifulSoup(response.text, 'html.parser')
      
        product_box = parser.findAll(class_="product-item")
        
        if (len(product_box) == 0):
            break

        for product in product_box:
            product_list.append(str(json.loads(product.get('data-view-content'))["click_data"]["id"]))
            

        i += 1
    return product_list, i

def save_product_id(product_list=[]):
    
    file = open(product_id_file, "w+")

    str = "\n".join(product_list)
    file.write(str)
    file.close()
    print("Save file: ", product_id_file)
def read_product_id():
    product_list=[]
    file = open(product_id_file, "r")
    while True:
        line= file.readline().strip()
        if not line:
         break
        product_list.append(line)
    file.close()
    return product_list  



```

```python
def crawl_product(product_list=[]):
    list_key=('id','name','thumbnail_url','description')
    output = pd.DataFrame()
    
    for product_id in product_list:
        response = requests.get(product_url.format(product_id),headers=headers)
        if (response.status_code == 200):
            product = {key: json.loads(response.text)[key] for key in json.loads(response.text) if key  in list_key}
            output = output.append(product, ignore_index=True)
            print(f"Crawl {typeStock}: ", product_id, ": ", response.status_code)
    return output
```

```python
# save product id for backup
product_list,list=crawl_product_id()
save_product_id(product_list)

# crawl detail for each product id
```

```python
product_list=read_product_id()
product_list = crawl_product(product_list)
```
